```python
import pandas as pd
from pandas import Series, DataFrame

hour_office  = month_office.iloc[:,2]
df_hour_office = Series(hour_office)
hour_apt  = month_apt.iloc[:,2]
df_hour_apt = Series(hour_apt)

def hourchange(hour):
    
    if   hour <= 5:
         hourRange = '0'
    
    elif hour <= 9 :
         hourRange ='1'
    
    elif hour <= 12 :
         hourRange = '2'
            
    elif hour <= 16 :
         hourRange = '3'
            
    elif hour <= 19 :
         hourRange = '4'
            
    elif hour <= 23 :
         hourRange = '5'
    else:
        hourRange =pd.np.nan
    return hourRange
            
```
